# Yakuza 0 Substories

```python
# load in the webpage
r = requests.get("https://yakuza.fandom.com/wiki/Yakuza_0/Substories")

# convert to beautiful soup object
soup = bs(r.content)
```

```python
table = soup.find_all("tbody")

# get kiryu's substories
kiryu_table = table[0]
columns = ["number", "name"]
rows = kiryu_table.find_all("tr")
l = []
for tr in rows:
    td = tr.find_all("td")
    row = [str(tr.get_text()).strip() for tr in td]
    l.append(row)
    
kiryu_y0_substories = pd.DataFrame(l, columns = columns)
kiryu_y0_substories = kiryu_y0_substories.dropna()
kiryu_y0_substories["character"] = "Kiryu Kazuma"
kiryu_y0_substories["game"] = "Yakuza 0"
kiryu_y0_substories = kiryu_y0_substories[["name", "character", "game"]]
kiryu_y0_substories["location"] = "Kamurocho"

# get majima's substories
majima_table = table[1]
majima_table
rows = majima_table.find_all("tr")
l = []
for tr in rows:
    td = tr.find_all("td")
    row = [str(tr.get_text()).strip() for tr in td]
    l.append(row)
    
majima_y0_substories = pd.DataFrame(l, columns = columns)
majima_y0_substories = majima_y0_substories.dropna()
majima_y0_substories["character"] = "Majima Goro"
majima_y0_substories["game"] = "Yakuza 0"
majima_y0_substories = majima_y0_substories[["name", "character", "game"]]
majima_y0_substories["location"] = "Sotenbori"

# combine the data frames
y0_substories = pd.concat([kiryu_y0_substories, majima_y0_substories])
y0_substories = y0_substories.reset_index(drop = True)
y0_substories
```

# About this data frame

This data frame contains the name of every substory within the mainline Yakuza games (Yakuza 0 - Yakuza: Like a Dragon).
For this data frame, I chose to use the substories from Yakuza Kiwami and Yakuza Kiwami 2 over the original versions of those
games released for the Playstation 2. In addition to the names, this data frame also has the character that does the substory, the game that the substory is in, and the location of the substory.

# Let's answer some questions

### Q1. What percent of the substories are done by Kiryu?

```python
total_substories = yakuza_substories["name"].count()
character_substories = yakuza_substories.groupby("character").count()["name"].reset_index()
total_kiryu = character_substories.loc[character_substories["character"] == "Kiryu Kazuma"].name
kiryu_percent = (total_kiryu / total_substories) * 100
kiryu_percent    # Answer = 66.83%

# Plot a pie chart to see the percentage done by every character
plt.style.use("default")
plt.title("Percentage of Substories Done by Each Character")
plt.pie(character_substories["name"], labels = character_substories["character"], autopct = "%.2f %%")
plt.show()

# I learned about value_counts after doing this whole analysis. That would have made some of my code a lot cleaner
# yakuza_substories["character"].value_counts()
```

```python
# Use a bar chart to see the exact number of substories done by each character

labels = character_substories["character"]
values = character_substories["name"]
plt.bar(labels, values)
plt.title("# of Substories Per Character")
plt.xticks(rotation = 90)
plt.xlabel("Character")
plt.ylabel("# of Substories")
# display the values on the each bar
for i in range(len(labels)):
        plt.text(i, values[i], values[i], ha = "center")
plt.show()
```

### Q2. What percentage of the substories happen in Kamurocho?

```python
locations = yakuza_substories.groupby("location").count()["name"].reset_index()
kamurocho = locations.loc[locations["location"] == "Kamurocho"]
kamurocho_percent = (kamurocho["name"] / total_substories) * 100
kamurocho_percent   # Answer = 55.17%

# Make a pie chart of the locations
labels = locations["location"]
values = locations["name"]
explode = (0, 0, .2, .5, .2, 0, 0, 0, 0)
plt.style.use("default")
plt.title("Percentage of Where Substories Take Place")
plt.pie(values, autopct = "%.2f %%", explode = explode, pctdistance = 0.99)
plt.legend(labels, loc = "upper center")
plt.show()
```

### Q3. How many substories contain the word "baby"?

```python
def has_baby(name):
    if ("baby" in name) or ("Baby" in name):
        return True
    else:
        return False
    
# Use the apply method and a custom function to create a new column in the data frame to make answering this question easy
df = yakuza_substories.copy(deep = True)
df["has_baby"] = df["name"].apply(lambda x: has_baby(x))
baby_in_name = df.loc[df["has_baby"]]
num_baby = baby_in_name.count()[0]    # Answer = 4
baby_in_name
```

### Q4. In which game did Kiryu have the most substories

```python
kiryu_games = yakuza_substories.loc[yakuza_substories["character"] == "Kiryu Kazuma"]
game_counts = kiryu_games.groupby("game").count().reset_index()
max_count = game_counts.loc[game_counts["name"] == game_counts["name"].max()]
max_count   # Answer = Yakuza 3

# Make a graph
game_counts = game_counts.reindex([0, 5, 6, 1, 2, 3, 4]) # Reorder the rows so that the games are in chronological order
game_counts
labels = game_counts["game"]
values = game_counts["name"]

plt.plot(labels, values, 'g.-')
plt.title("# of Substories Kiryu has in each Game")
plt.ylabel("# of Substories")
plt.xlabel("Game")
plt.xticks(rotation = 90)
plt.show()
```

### Q5. What characters do not have any substories in Kamurocho?

```python
c_location = yakuza_substories[["character", "location"]]

# get all the unique character and location combinations
c_location = c_location.drop_duplicates()

# get list of all characters
charas = c_location["character"].unique()

# get all characters who do have substories in Kamurocho
in_kamurocho = c_location.loc[c_location["location"] == "Kamurocho"]
kamurocho_charas = in_kamurocho["character"].to_list()

# compare the two lists to get our answer (Use XOR)
not_in_kamurocho = set(charas) ^ set(kamurocho_charas)
not_in_kamurocho  # Answer = Kasuga Ichiban, Majima Goro, and Shinada Tatsuo
```

### Q6. How many games does each character appear in (as a playable character)?

```python
# size returns the number of unique groups
chara_game = yakuza_substories.groupby(["character", "game"]).size().reset_index()
game_counts = chara_game.groupby("character").count()["game"]
game_counts
```
